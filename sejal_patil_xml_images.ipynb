```python
# Downloading using wget https://stackoverflow.com/a/14456285/9292995
# Downloading using requests https://www.codingem.com/python-download-file-from-url/
import requests

def download_url(URL, file_name):
    
    # 2. download the data behind the URL
    response = requests.get(URL)
    # 3. Open the response into a new file called instagram.ico
    open(file_name, "wb").write(response.content)


URL = "https://registers.esma.europa.eu/solr/esma_registers_firds_files/select?q=*&fq=publication_date:%5B2021-01-17T00:00:00Z+TO+2021-01-19T23:59:59Z%5D&wt=xml"
file_name = "data.xml"
download_url(URL, file_name)
```

```python
import csv
import xml.etree.ElementTree as ET
```

```python
# create element tree object
tree = ET.parse('data.xml')

# get root element
root = tree.getroot()

for r in root:
    for item in r:
        for element in item:
            if element.text == 'DLTINS':
                download_link = item[1].text
                break  # breaking after getting first download link

print(f"First link of file_type DLTINS is {download_link}.")
print(f"Saving {download_link} as data.zip")
download_url(download_link, 'data.zip')
```

```python
# https://www.geeksforgeeks.org/convert-xml-to-csv-in-python/
# Importing the required libraries
import xml.etree.ElementTree as Xet
import pandas as pd

cols = ["FinInstrmGnlAttrbts.Id", 
        "FinInstrmGnlAttrbts.FullNm", 
        "FinInstrmGnlAttrbts.ClssfctnTp", 
        "FinInstrmGnlAttrbts.CmmdtyDerivInd", 
        "FinInstrmGnlAttrbts.NtnlCcy",
        "Issr"]
rows = []

# Parsing the XML file
xmlparse = Xet.parse('data/'+file_name)
root = xmlparse.getroot()
for i in root:
    Id = i.find(cols[0]).text
    FullNm = i.find(cols[1]).text
    ClssfctnTp = i.find(cols[2]).text
    CmmdtyDerivInd = i.find(cols[3]).text
    NtnlCcy = i.find(cols[4]).text
    Issr = i.find(cols[5]).text
    
    rows.append({cols[0]: Id,
                 cols[1]: FullNm,
                 cols[2]: ClssfctnTp,
                 cols[3]: CmmdtyDerivInd,
                 cols[4]: NtnlCcy,
                 cols[5]: Issr})

df = pd.DataFrame(rows, columns=cols)

# Writing dataframe to csv
df.to_csv('output.csv')
```
